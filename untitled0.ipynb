```python
import gspread
from google.oauth2.service_account import Credentials
from google.colab import files
import json
import datetime

# ID твоей таблицы
SPREADSHEET_ID = "1ys1rYj6iNMMAm2w1VgQMYdOtYZRv9-jbkxfqQheAFso"

# Шаг 1: Загружаем JSON-файл сервисного аккаунта
print("📁 Загрузите JSON-файл сервисного аккаунта:")
uploaded = files.upload()

# Шаг 2: Преобразуем загруженные байты в словарь
SERVICE_ACCOUNT_INFO = None
for filename in uploaded:
    # uploaded[filename] — это bytes, нужно декодировать в строку, потом превратить в словарь
    json_str = uploaded[filename].decode('utf-8')  # bytes → строка
    SERVICE_ACCOUNT_INFO = json.loads(json_str)     # строка → словарь
    print(f"✅ Загружен и преобразован: {filename}")

# Проверка, что получился словарь
print(f"Тип SERVICE_ACCOUNT_INFO: {type(SERVICE_ACCOUNT_INFO)}")  # Должно быть <class 'dict'>

def init_google_sheets():
    """Инициализация подключения к Google Sheets"""
    try:
        # Используем загруженные данные сервисного аккаунта (теперь это dict!)
        credentials = Credentials.from_service_account_info(
            SERVICE_ACCOUNT_INFO,
            scopes=['https://www.googleapis.com/auth/spreadsheets']
        )
        gc = gspread.authorize(credentials)
        sheet = gc.open_by_key(SPREADSHEET_ID).sheet1
        print("✅ Подключение к Google Sheets успешно!")
        return sheet
    except Exception as e:
        print(f"❌ Ошибка подключения: {e}")
        return None

# Создаём глобальный объект листа
sheet = init_google_sheets()

async def write_to_google_sheets(user_name: str, phone: str, tour_name: str, user_id: int):
    """Асинхронная запись данных в Google Sheets"""
    try:
        now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        row = [now, user_name, phone, tour_name, str(user_id)]

        if sheet:
            sheet.append_row(row)
            print(f"📊 Данные записаны в Sheets: {row}")
            return True
        else:
            print("❌ Нет подключения к Sheets")
            return False
    except Exception as e:
        print(f"❌ Ошибка записи: {e}")
        return False
```

```python
!pip install nest_asyncio --upgrade -q
import asyncio
import json
import datetime
from typing import Dict, Tuple

from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import Application, CommandHandler, CallbackQueryHandler, ContextTypes, MessageHandler, filters
import nest_asyncio

# Apply nest_asyncio to allow running asyncio loops inside existing ones
nest_asyncio.apply()

# ===== НАСТРОЙКИ =====
# Твой токен от @BotFather
TOKEN = "8647241736:AAF4Y5McqUsL1E4uvUI4xaENHSXCrFf0ToY"

# Ссылка на ваш сайт (GitHub Pages)
SITE_URL = "https://selkaterin.github.io/my-website/"

print("✅ Библиотеки загружены. Настройки готовы.")

# Глобальный словарь для хранения данных о пользователях, записавшихся
user_bookings = {}

# ===== КНОПКИ С ЭМОДЗИ =====
# Словарь соответствия туров и эмодзи
emoji_map = {
    "Озеро Рица": "🔵 Озеро Рица",
    "Новоафонская пещера": "🕯️ Новоафонская пещера",
    "Новоафонский монастырь": "⛪ Новоафонский монастырь",
    "Анакопийская крепость": "🏰 Анакопийская крепость",
    "Гагрская колоннада": "🏛️ Гагрская колоннада",
    "Юпшарский каньон": "🏞️ Юпшарский каньон",
    "Источники Къндыг": "💧 Источники Къндыг",
    "Дача Сталина": "🏚️ Дача Сталина",
    "Ботанический сад": "🟢 Ботанический сад",
    "Хашуписанский каньон": "🏞️ Хашуписанский каньон"
}

# Функция создания главного меню (список туров с эмодзи)
def get_main_keyboard() -> InlineKeyboardMarkup:
    """Создаёт клавиатуру с кнопками туров с эмодзи"""
    keyboard = []
    # Разбиваем туры на строки по 2 кнопки
    for i in range(0, len(tour_list), 2):
        row = []
        # Первая кнопка в строке
        tour_name = tour_list[i]
        row.append(InlineKeyboardButton(
            emoji_map.get(tour_name, tour_name),  # ← с эмодзи
            callback_data=f"tour_{tour_name}"
        ))
        # Вторая кнопка в строке (если есть)
        if i + 1 < len(tour_list):
            tour_name2 = tour_list[i + 1]
            row.append(InlineKeyboardButton(
                emoji_map.get(tour_name2, tour_name2),  # ← с эмодзи
                callback_data=f"tour_{tour_name2}"
            ))
        keyboard.append(row)

    # Добавляем отдельную строку с кнопкой "Наш сайт" и "Записаться"
    keyboard.append([
        InlineKeyboardButton("🌐 Наш сайт", url=SITE_URL),
        InlineKeyboardButton("📝 Записаться на тур", callback_data="booking_menu")
    ])

    return InlineKeyboardMarkup(keyboard)

# Обработчик команды /start
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    """Приветствие и главное меню"""
    welcome_text = (
        "🏝️ **Добро пожаловать в турагентство Абхазии!**\n\n"
        "Я помогу вам выбрать экскурсию по самым живописным местам.\n"
        "Выберите направление из списка ниже 👇\n\n"
        "⚠️ *Для записи нажмите «Записаться на тур»*"
    )
    await update.message.reply_text(
        welcome_text,
        reply_markup=get_main_keyboard(),
        parse_mode="Markdown"
    )

# Обработчик нажатия на кнопки (callback_query)
async def button_handler(update: Update, context: ContextTypes.DEFAULT_TYPE):
    query = update.callback_query
    await query.answer()

    data = query.data

    # Если нажали на тур
    if data.startswith("tour_"):
        tour_name = data.replace("tour_", "")
        tour_info = tours_data.get(tour_name)

        if tour_info:
            message_text = (
                f"🏔️ **{tour_name}**\n\n"
                f"{tour_info['description']}\n\n"
                f"📞 **Связь с гидом:** {tour_info['guide']}\n"
                f"⏱️ Длительность: {tour_info['duration']}\n\n"
                f"👇 Чтобы записаться, нажмите кнопку ниже:"
            )

            keyboard = InlineKeyboardMarkup([[
                InlineKeyboardButton(
                    "✅ Записаться на этот тур",
                    callback_data=f"book_{tour_name}"
                ),
                InlineKeyboardButton("🔙 Назад к списку", callback_data="back_to_menu")
            ]])

            await query.edit_message_text(
                message_text,
                reply_markup=keyboard,
                parse_mode="Markdown"
            )
        else:
            await query.edit_message_text("❌ Тур не найден. Попробуйте позже.")

    # Если нажали "Назад к списку"
    elif data == "back_to_menu":
        await query.edit_message_text(
            "🏝️ **Выберите направление экскурсии:**",
            reply_markup=get_main_keyboard(),
            parse_mode="Markdown"
        )

    # Если нажали "Записаться на тур" из главного меню
    elif data == "booking_menu":
        keyboard = []
        for tour in tour_list:
            keyboard.append([InlineKeyboardButton(
                f"📌 {tour}",
                callback_data=f"book_{tour}"
            )])
        keyboard.append([InlineKeyboardButton("🔙 Назад", callback_data="back_to_menu")])

        await query.edit_message_text(
            "📋 **Выберите тур для записи:**\n"
            "После выбора я попрошу ваше имя и телефон.",
            reply_markup=InlineKeyboardMarkup(keyboard),
            parse_mode="Markdown"
        )

    # Если нажали "Записаться на этот тур" из описания тура
    elif data.startswith("book_"):
        tour_name = data.replace("book_", "")
        context.user_data['booking_tour'] = tour_name

        keyboard = InlineKeyboardMarkup([[
            InlineKeyboardButton("🔙 Отмена", callback_data="back_to_menu")
        ]])

        await query.edit_message_text(
            f"✍️ **Запись на тур:** {tour_name}\n\n"
            f"Введите ваше **имя** (или нажмите отмена):\n\n"
            f"*(Просто напишите сообщение с вашим именем)*",
            reply_markup=keyboard,
            parse_mode="Markdown"
        )

        context.user_data['awaiting_name'] = True

# Обработчик текстовых сообщений (для ввода имени и телефона)
async def text_handler(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_id = update.effective_user.id
    text = update.message.text

    # Если ожидаем имя
    if context.user_data.get('awaiting_name'):
        context.user_data['user_name'] = text
        context.user_data['awaiting_name'] = False
        context.user_data['awaiting_phone'] = True

        await update.message.reply_text(
            f"Приятно познакомиться, {text}! 📞\n\n"
            "Теперь отправьте ваш **номер телефона** (в любом формате):\n\n"
            "*(Например: +7 999 123 45-67 или просто 89991234567)*",
            parse_mode="Markdown"
        )
        return

    # Если ожидаем телефон
    if context.user_data.get('awaiting_phone'):
        phone = text
        tour_name = context.user_data.get('booking_tour', 'не выбран')
        user_name = context.user_data.get('user_name', 'неизвестно')

        # Сохраняем данные в словарь
        booking_id = f"{user_id}_{datetime.datetime.now().timestamp()}"
        user_bookings[booking_id] = {
            'user_id': user_id,
            'user_name': user_name,
            'phone': phone,
            'tour': tour_name,
            'datetime': datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }

        # Сбрасываем состояния
        context.user_data['awaiting_phone'] = False
        context.user_data['user_name'] = None
        context.user_data['booking_tour'] = None

        # Запись в Google Sheets
        try:
            await write_to_google_sheets(user_name, phone, tour_name, user_id)
            print(f"📝 Новая запись в Sheets: {user_name}, {phone}, {tour_name}")
        except Exception as e:
            print(f"❌ Ошибка записи в Google Sheets: {e}")

        # Подтверждение записи
        await update.message.reply_text(
            f"✅ **Вы успешно записаны!**\n\n"
            f"📌 Тур: *{tour_name}*\n"
            f"👤 Имя: {user_name}\n"
            f"📞 Телефон: {phone}\n\n"
            f"С вами свяжется гид для подтверждения.\n"
            f"А пока можете посмотреть наш сайт: {SITE_URL}",
            parse_mode="Markdown",
            reply_markup=get_main_keyboard()
        )
        return

    # Если ничего не ожидаем — даём подсказку
    await update.message.reply_text(
        "Используйте кнопки меню для навигации. Чтобы начать заново, нажмите /start",
        reply_markup=get_main_keyboard()
    )

# Главная функция запуска
async def main():
    print("🚀 Запуск бота...")
    app = Application.builder().token(TOKEN).build()

    # Регистрируем обработчики
    app.add_handler(CommandHandler("start", start))
    app.add_handler(CallbackQueryHandler(button_handler))
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, text_handler))

    print("✅ Бот готов! Начинаем поллинг...")
    await app.run_polling(close_loop=False)

# Точка входа для Colab
try:
    await main()
except SystemExit:
    print("Bot stopped gracefully due to SystemExit.")
```
